# Words

```python
sns.barplot(y=train['keyword'].value_counts()[:20].index,x=train['keyword'].value_counts()[:20])
```

Where the keyword is fatalities:

```python
def bar_plot_for_word(word: str):
    sns.countplot(data=train[train['keyword'] == word], x='target')
```

```python
bar_plot_for_word('fatalities')
```

If label is true, word count:

```python
df_disaster = train[train['target'] == 1]
sns.barplot(y=df_disaster['keyword'].value_counts()[:20].index,x=df_disaster['keyword'].value_counts()[:20])
```

```python
bar_plot_for_word('derailment')
```

```python
bar_plot_for_word('wreckage')
```

No disaster

```python
df_no_disaster = train[train['target'] == 0]
sns.barplot(y=df_no_disaster['keyword'].value_counts()[:20].index,x=df_no_disaster['keyword'].value_counts()[:20])
```

```python
bar_plot_for_word('body%20bags')
```

```python
bar_plot_for_word('harm')
```
